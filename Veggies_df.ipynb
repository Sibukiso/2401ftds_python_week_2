Let check at price trend of each vegitable in a year and what time of the year each vegitable price is at the maximum

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

# Load veggies csv dataset
Veggies_df = pd.read_csv('C:\\Users\\sibuk\\Documents\\Team-J3-Python-final-project\\Veggies_df.csv')

# Convert 'Price Dates' column to datetime with the correct format
Veggies_df['Price Dates'] = pd.to_datetime(Veggies_df['Price Dates'], format='%d-%m-%Y')

# Set 'Price Dates' column as index
Veggies_df.set_index('Price Dates', inplace=True)

# Plotting the prices of each vegetable over time
plt.figure(figsize=(12, 8))
for column in Veggies_df.columns:
    sns.lineplot(data=Veggies_df, x=Veggies_df.index, y=column, label=column)

plt.title('Trend Analysis of Vegetable Prices')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(title='Vegetable')
plt.show()
```

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt

# Load veggies csv dataset
Veggies_df = pd.read_csv('C:\\Users\\sibuk\\Documents\\Team-J3-Python-final-project\\Veggies_df.csv')

# Convert Price Dates column to datetime with the correct format
Veggies_df['Price Dates'] = pd.to_datetime(Veggies_df['Price Dates'], format='%d-%m-%Y')

# Setting Price Dates column as index
Veggies_df.set_index('Price Dates', inplace=True)

# Plotting the prices of each vegetable separately
plt.figure(figsize=(12, 8))
for column in Veggies_df.columns:
    plt.figure(figsize=(8, 4))  # Create a new figure for each vegetable
    sns.lineplot(data=Veggies_df, x=Veggies_df.index, y=column)
    max_price_date = Veggies_df.index[Veggies_df[column] == Veggies_df[column].max()][0]
    max_price_value = Veggies_df[column].max()
    plt.axvline(x=max_price_date, color='red', linestyle='--', label=f'Max Price: {max_price_value}')
    plt.title(f'Trend Analysis of {column} Prices')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend([f'Max Price: {max_price_value}\nDate: {max_price_date.strftime("%Y-%m-%d")}'], loc='upper left')
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load veggies csv dataset
Veggies_df = pd.read_csv('C:\\Users\\sibuk\\Documents\\Team-J3-Python-final-project\\Veggies_df.csv')

# Iterate over each vegetable and plot its violin plot
for vegetable in Veggies_df.columns:
    if vegetable != 'Price Dates':  # Skip plotting for 'Price Dates' column
        # Create a figure and axis for the plot
        plt.figure(figsize=(8, 6))
        ax = sns.violinplot(x=Veggies_df[vegetable], color='skyblue')
        
        # Set title and labels
        ax.set_title(f'Violin Plot of {vegetable} Prices')
        ax.set_xlabel('Price')
        ax.set_ylabel('Density')
        
        # Show the plot
        plt.show()
```
